```python
x=[1,2,3,4,5]
y=[5,4,2,1,3]
# plt.scatter(x,y)
fig=plt.figure(figsize=(3,3))
plt.plot(x,y)
plt.xlabel('x list')
plt.ylabel('y list')
plt.xlim(1,8)
plt.ylim(1,8)
plt.title('line plot')
plt.suptitle('Sales',size=5)
```
